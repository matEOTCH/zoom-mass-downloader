```python
# 1. Instalación de herramientas
!pip install yt-dlp pandas
from google.colab import drive, files
import pandas as pd
import os
import io
import time

# 2. Conectar con Google Drive
print("🔗 Conectando a Google Drive...")
drive.mount('/content/drive')

# 3. INTERFAZ: Nombre de carpeta
print("\n" + "="*40)
nombre_carpeta_drive = input("👉 ¿Cómo se llamará la carpeta en tu Drive?: ")
print("="*40 + "\n")

# 4. INTERFAZ: Subir archivo
print("📁 Por favor, sube tu archivo (.csv o .txt) con los links:")
subido = files.upload()
nombre_archivo_subido = list(subido.keys())[0]

# 5. CONFIGURACIÓN DE RUTAS
RUTA_DRIVE = f'/content/drive/MyDrive/{nombre_carpeta_drive}'
if not os.path.exists(RUTA_DRIVE):
    os.makedirs(RUTA_DRIVE)
    print(f"\n✨ Carpeta creada: '{nombre_carpeta_drive}'")
else:
    print(f"\n✅ Usando carpeta existente: '{nombre_carpeta_drive}'")

# 6. FUNCIÓN DE DESCARGA
def descargar_zoom_pro(url, nombre_clase):
    # Limpiar nombre de archivo
    nombre_archivo = "".join([c for c in nombre_clase if c.isalnum() or c in (' ', '_')]).rstrip()
    ruta_final = os.path.join(RUTA_DRIVE, f"{nombre_archivo}.mp4")

    if os.path.exists(ruta_final):
        print(f"⏭️  Saltando: '{nombre_clase}' (Ya existe)")
        return

    print(f"📥 Descargando: {nombre_clase}...")

    # Comando yt-dlp
    command = f'yt-dlp -o "{ruta_final}" "{url}"'
    resultado = os.system(command)

    if resultado == 0:
        print(f"✅ ¡Éxito!")
    else:
        print(f"⚠️ Error en '{nombre_clase}'. Verifica si el link pide contraseña.")

# 7. PROCESAMIENTO DEL ARCHIVO
try:
    # Intentamos detectar si es CSV (coma) o TXT (tabulación como en tu captura)
    contenido = subido[nombre_archivo_subido].decode('utf-8')
    separador = '\t' if '\t' in contenido else ','

    df = pd.read_csv(io.StringIO(contenido), header=None, sep=separador)

    print(f"\n🚀 Iniciando descarga de {len(df)} videos...\n")

    for index, fila in df.iterrows():
        nombre_video = str(fila[0])
        link_video = str(fila[1])

        if "http" in link_video:
            descargar_zoom_pro(link_video, nombre_video)
            time.sleep(1)
        else:
            print(f"❌ Fila {index+1}: Link no reconocido.")

except Exception as e:
    print(f"❌ Error al leer el archivo: {e}")

print("\n" + "="*40)
print("🏁 PROCESO FINALIZADO")
print(f"Tus videos están en Drive: /MyDrive/{nombre_carpeta_drive}")
print("="*40)
```
